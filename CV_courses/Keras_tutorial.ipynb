```python
from numpy import loadtxt  # Fixed typo
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense  # Fixed typo
```

```python
# load the dataset
import pandas as pd
pd.read_csv('pima-indians-diabetes.data.csv')
dataset = loadtxt('pima-indians-diabetes.data.csv', delimiter=',')
# split into input (X) and output (y) variables
X = dataset[:,0:8]
y = dataset[:,8]
```

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv')
```

```python
#define the keras model
model = Sequential()
model.add(Dense(30, input_shape = (8,), activation = 'relu'))
model.add(Dense(20, activation ='relu'))
model.add(Dense(15, activation ='relu'))
model.add(Dense(12, activation ='relu'))
model.add(Dense(9, activation ='relu'))
model.add(Dense(7, activation ='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
#compile the keras model
model.compile(loss ='binary_crossentropy', optimizer ='adam', metrics = ['accuracy'])
```

```python
#fit the keras model on the dateset
model.fit(X, y , epochs =100, batch_size =10)
```

```python
#evaluate the keras model
_, accuracy = model.evaluate(X,y)
print('Accuracy: %.2f' % (accuracy * 100))
```
